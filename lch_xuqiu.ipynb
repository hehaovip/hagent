```python
from langchain.prompts import PromptTemplate
from langchain.llms import OpenAI
from langchain.chains import LLMChain
from langchain_openai import ChatOpenAI

product = "商城软件"
ans = ['用户管理']

llm = ChatOpenAI(temperature=0.9,
    model="glm-4",
    openai_api_key="964ae3f9c2f746a3c707037ec9466b15.7hbNv7CyS6lVGQWg",
    openai_api_base="https://open.bigmodel.cn/api/paas/v4/")

prompt = PromptTemplate(
    input_variables=["product", "name"],
    template="{product}的功能模块包括{name}这些，额外在写出一个以上描述中不存在的 {product}的功能模块，记住只回复这个功能模块的名称。",
)

prompt2 = PromptTemplate(
    input_variables=["product", "name"],
    template="描述{product}的{name}功能模块的详细开发逻辑，越详细越好，以便开发者快速理解需求并实施开发。",
)

prompt3 = PromptTemplate(
    input_variables=["product", "name"],
    template="根据{product}描述的{name}开发逻辑，一步一步的写出php的thinkphp框架开发代码。",
)

chain1 = LLMChain(llm=llm, prompt=prompt)
chain2 = LLMChain(llm=llm, prompt=prompt2)
chain3 = LLMChain(llm=llm, prompt=prompt3)

for i in range(10):
    
    ff = chain1.run({
        'product': product,
        'name': ans
        })
    ff2 = chain2.run({
        'product': product,
        'name': ff
        })
    ff3 = chain3.run({
        'product': ff,
        'name': ff2
        })
    ans.append(str(ff))
    print(ff)
    print('123')
    print(ff2)
    print('code')
    print(ff3)
    print('321')
# ans.apeend(ff)
# 仅指定输入变量运行链式连接。
print("123")
```
